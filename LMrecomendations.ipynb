```python
#importing libraries
import numpy as np
import pandas as pd
import neattext.functions as nfx
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('course_data.csv')
df.head()
```

```python
df['Clean_title'] = df['course_title'].apply(nfx.remove_stopwords) #removing stopwords from course titles
df['Clean_title'] = df['Clean_title'].apply(nfx.remove_special_characters) #removing special characters

df.head()
```

```python
#vectorizig clean title
countvect = CountVectorizer()
cvmat = countvect.fit_transform(df['Clean_title'])
cvmat
```

```python
#finding vector similarities
cos_sim = cosine_similarity(cvmat)
cos_sim.shape
cos_sim
```

```python
title = 'How To Maximize Your Profits Trading Options'
course_index = pd.Series(df.index, index = df['course_title']).drop_duplicates()
#getting indexes
index = course_index[title]

#finding vector similarities
cos_sim[index]
```

```python
scores = list(enumerate(cos_sim[index]))
#scores
```

```python
sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True) #sorting from the highest match
#sorted_scores
```

```python
#getting sorted index and similarity from similarity list
index_sorted = [x[0] for x in sorted_scores]
similarity_sorted = [x[1] for x in sorted_scores]
```

```python
#creating recommendation df based on
recommendation_df = df.iloc[index_sorted]
recommendation_df
```

```python
#adding similarity scores column
recommendation_df['similarity_score'] = similarity_sorted
recommendation_df
```

```python
recommendation_df = recommendation_df[['course_title', 'similarity_score', 'price', 'num_subscribers']]
recommendation_df = recommendation_df[recommendation_df['similarity_score'] > 0.5]
```
